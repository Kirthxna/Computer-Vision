```python
import math 
def buildMat(s):
    Arr = s.split()
    A = []
    x = 0
    for i in range(3):
        inte = []
        for j in range(3):
            inte.append(int(Arr[x]))
            x = x+1
        A.append(inte)
    return A


A = buildMat(input())
B = buildMat(input())

#SSD BEGIN
ssd = 0
for i in range(3):
    for j in range(3):
        dif = A[i][j] - B[i][j]
        print(str.format("{0} - {1} = {2}", A[i][j], B[i][j], dif))
        ssdin = dif ** 2
        print(str.format("SSD -> {} , {}  = {} ", A[i][j], B[i][j], ssdin))
        ssd += ssdin

print("SSD - {0}", ssd)

#SSDEND

#NCCBEGIN

sumG = 0
sumImg = 0

for i in range(3):
    for j in range(3):
        sumG += B[i][j]
        sumImg += A[i][j]
meanG = sumG / 9
meanImg = sumImg / 9

print(str.format("meanG - {0}", meanG))
print(str.format("meanImg - {0}", meanImg))


numSum = 0
denG = 0
denImg = 0
for i in range(3):
    for j in range(3):
        num = (A[i][j] - meanImg) * (B[i][j] - meanG)
        print(str.format("f{3}{4} -> {0} - {1} = {2}", A[i][j] , meanImg, (A[i][j] - meanImg), i, j))
        print(str.format("g{3}{4} -> {0} - {1} = {2}", B[i][j] , meanG, (B[i][j] - meanG), i, j))
        print(str.format(str.format("f{0}{1} * g{0}{1} = {2}", i, j , num)))
        numSum += num
        denG += (B[i][j] - meanG) ** 2
        print(str.format("Den-G{1}{2} = {0}", denG, i, j))
        denImg += (A[i][j] - meanImg) ** 2
        print(str.format("Den-F{1}{2} = {0}", denImg, i, j))
```
